```python
# cv2.calcHist()  
# BGR  HSV(  BGR  HSV)
# 2D  :
# channels=[01]  H  S 
# bins=[180256]H  180S  256
# range=[01800256]H  0  180S   0  256

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/Users/ylshan/Documents/python//opencv/data/lion.jpeg')
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
hist = cv2.calcHist([hsv], [0, 1], None, [180, 256], [0, 180, 0, 256])

plt.imshow(img)
plt.show() 
plt.imshow(hist,interpolation = 'nearest')
plt.show() 
```
